## Regresión Lineal en Machine Learning 1

Crea un DataFrame de Pandas, llamado `df` con dos columnas: `"Edad"` y `"Salario"`.

La columna `"Edad"` debe tener los valores [25, 30, 35, 40, 45]

La columna `"Salario"` debe tener los valores [3000, 3500, 4000, 4500, 5000].

Asegurate de cumplir exactamente con los puntos siguientes:

- Crea un diagrama de dispersión `scatter` para visualizar la relación entre la edad y el salario.
- El gráfico debe tener puntos (`'o'`) azules (`blue`).
- Titulo del gráfico: `'Relación entre Edad y Salario'`
- Etiqueta del eje x: `'Edad'`
- Etiqueta del eje y: `'Salario'`
- Activa la rejilla en el gráfico (grid)

Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.

**Observación:**

El gráfico que debes obtener será identico a este:

<img src="assets/img/1 regresion_lineal_ejercicio_1.png" width="300px">

````python
import pandas as pd
import matplotlib.pyplot as plt
 
# Crear DataFrame
data = {'Edad': [25, 30, 35, 40, 45],
        'Salario': [3000, 3500, 4000, 4500, 5000]}
df = pd.DataFrame(data)
 
# Crear el diagrama de dispersión
plt.scatter(df['Edad'], df['Salario'], color='blue', marker='o')
plt.title('Relación entre Edad y Salario')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid(True);
````

## Regresión Lineal en Machine Learning 2

Utilizando los datos del ejercicio anterior:

```python
# Crear DataFrame
data = {'Edad': [25, 30, 35, 40, 45],
        'Salario': [3000, 3500, 4000, 4500, 5000]} 
df = pd.DataFrame(data)
```

Crea un modelo de regresión lineal para predecir el salario basándose en la edad.

Asegurate de mantener la información del gráfico anterior:

- Titulo del gráfico: `'Relación entre Edad y Salario'`
- Etiqueta del eje x: `'Edad'`
- Etiqueta del eje y: `'Salario'`
- Activa la rejilla en el gráfico (grid)

**Nota: No debes crear el gráfico de dispersión, solamente debes crear el modelo de regresión lineal.**

Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.

**Observación:**

El gráfico que debes obtener será identico a este:

<img src="assets/img/1 regresion_lineal_ejercicio_2.png" width="300px">

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
 
# Crear DataFrame
data = {'Edad': [25, 30, 35, 40, 45],
        'Salario': [3000, 3500, 4000, 4500, 5000]}
df = pd.DataFrame(data)
 
# Preparar los datos para la regresión
X = df[["Edad"]] 
y = df["Salario"]     
 
# Crear y ajustar el modelo de regresión lineal
modelo = linear_model.LinearRegression()
modelo.fit(X, y)
 
# Crear el gráfico
plt.title('Relación entre Edad y Salario')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid(True)
plt.plot(df["Edad"],
        modelo.predict(X));
````
