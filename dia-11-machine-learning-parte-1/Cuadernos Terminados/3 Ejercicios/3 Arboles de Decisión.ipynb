## Árboles de Decisión - Parte I (2)

Utiliza el conjunto de datos "iris" cargado previamente en el dataframe

```python
from iris_info import iris_dict
# Convertir el diccionario a DataFrame
iris_df = pd.DataFrame(iris_dict)
```

Realiza los siguientes pasos:

1. Prepara el DataFrame para modelado, eliminando la columna `'species'` y usando el resto como características (Almacénalo en la variable `X`).
2. Inicializa el codificador de etiquetas y transforma las etiquetas categóricas de especies en valores numéricos. (Almacénalo en la variable `y`). Utiliza `LabelEncoder` de scikit-learn para transformar las etiquetas de especies de texto a valores numéricos. Esto es necesario para entrenar el modelo de árbol de decisión.

Visualización:

- Genera un gráfico de dispersión (`scatter`) para visualizar la relación entre la longitud del sépalo y las especies codificadas `X` y `y`.
- Utiliza `c=y`, para colorear los puntos según la codificación de especies.
- Añade una barra de colores al gráfico para indicar qué color corresponde a cada especie codificada. La barra debe tener el siguiente título: `Especies codificadas`
- Título del gráfico: `'Relación entre la longitud del sépalo y las especies codificadas'`
- Eje x: `'Longitud del Sépalo (cm)'`
- Eje y:`'Especies codificadas'`
- Activa la rejilla del gráfico (grid)

Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.

**Observación:**

El gráfico que debes obtener será idéntico a este:

<img src="assets/img/3 arboles_decision_parte_1_ejercicio_2.png" width="300px">

````python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
 
# Convertir el diccionario que contiene los datos del dataset iris en un DataFrame de pandas para facilitar su manipulación.
iris_df = sns.load_dataset("iris")
 
# Preparar el DataFrame para modelado, eliminando la columna 'species' y usando el resto como características (X).
X = iris_df.drop("species", axis=1)
 
# Inicializar el codificador de etiquetas y transformar las etiquetas categóricas de especies en valores numéricos.
le = LabelEncoder()
especies = le.fit_transform(iris_df["species"])
y = especies
 
# Configuración de la figura y creación de un gráfico de dispersión para visualizar la relación
# entre la longitud del sépalo (una característica de las flores iris) y las especies codificadas.
plt.scatter(X['sepal_length'], y, c=y)  # Usar color amarillo para una mejor visualización
 
# Añadir una barra de colores que representa las diferentes especies codificadas para interpretar mejor el gráfico.
plt.colorbar(label='Especies codificadas')
 
# Configuración de los títulos de los ejes y el título principal del gráfico.
plt.title('Relación entre la longitud del sépalo y las especies codificadas')
plt.xlabel('Longitud del Sépalo (cm)')
plt.ylabel('Especies codificadas')
 
# Habilitar la cuadrícula en el gráfico para mejorar la legibilidad de los datos visualizados.
plt.grid(True);
````

## Árboles de Decisión - Parte I (3)

Entrena y visualiza el árbol de decisión con los datos del Ejercicio 2

Completa el código proporcionado para lograr los siguientes pasos:

- Divide los datos en conjuntos de entrenamiento y de prueba
  `X_entrena, X_prueba, y_entrena, y_prueba`
  La división entre entrenamiento y prueba debe estar en proporción (80% para entrenamiento, 20% para prueba). Asigna el `random_state` que desees.

- Crea y entrena al modelo del árbol de decisión. `X_entrena, y_entrena`

- Visualización del árbol de decisión con un tamaño de `20, 20`

Recuerda importar los módulos de python necesarios para que tu código funcione correctamente.

**Observación:**

El gráfico que debes obtener será idéntico a este:

<img src="assets/img/3 arboles_decision_parte_1_ejercicio_3.png" width="300px">

````python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
 
#___________IMPORTA AQUI LAS LIBRERIAS QUE REQUIERAS____________
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
 
# Convertir el diccionario que contiene los datos del dataset iris en un DataFrame de pandas para facilitar su manipulación.
iris_df = sns.load_dataset("iris")
 
# Preparar el DataFrame para modelado, eliminando la columna 'species' y usando el resto como características (X).
X = iris_df.drop("species", axis=1)
 
# Inicializar el codificador de etiquetas y transformar las etiquetas categóricas de especies en valores numéricos.
le = LabelEncoder()
especies = le.fit_transform(iris_df["species"])
y = especies
 
#___________COMPLETA EL CÓDIGO CON LOS SIGUIENTES PASOS____________
 
# División de datos en conjunto de entrenamiento y de prueba
X_entrena, X_prueba, y_entrena, y_prueba = train_test_split(X, y, train_size=0.8, random_state=42)
 
# Creación y entrenamiento del modelo de árbol de decisión
arbol = DecisionTreeClassifier()
arbol.fit(X_entrena, y_entrena)
 
# Visualización del árbol de decisión
plt.figure(figsize=(20, 20))
plot_tree(decision_tree=arbol);
````
