# Gráficos Pastel (Pie Charts)

Pocos gráficos son tan populares y tan fáciles de interpretar como los gráficos circulares, también conocidos como *gráficos de torta*, o *gráficos pastel*, o en inglés *pie charts*. Por supuesto que no los podíamos dejar fuera de estas explicaciones.

Aquí tenemos un pequeño DataFrame con información sobre cantidades de frutas en un almacén, con el que vamos a prácticar nuestras habilidades para crear gráficos de torta.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.DataFrame({
    "etiquetas":['Manzanas', 'Bananas', 'Naranjas', 'Uvas'],
    "cantidad":[30, 60, 80, 20]
})
df
```

Vamos a representar estas cantidades en un gráfico de torta. Para eso contamos con el método `pie()` de Matplotlib que significa *pastel*.

Hacer un gráfico circular es muy sencillo, porque el único parámetro que nos pide `pie()` es *x*, es decir que solo nos pide una **secuencia de valores**.

En nuestro caso, será suficiente con que le pasemos una **Serie** de nuestro DataFrame.

```python
help(plt.pie)
```

```python
plt.pie(df['cantidad']);
```

Mira con qué poco esfuerzo ya tenemos un vistoso y colorido gráfico.

Por supuesto que aquí podemos configurar muchas cosas para que además de práctico sea informativo. Entonces nos vamos a meter con sus **parámetros** para ver algunas de las muchas cosas que podemos hacer aquí.

Comencemos por agregar etiquetas que nos permitan identificar a qué grupo de frutas pertenece cada porción de nuestor pastel. Para eso `pie()` cuenta con el atributo `labels`, al que le podemos pasar una lista de **etiquetas**, o mejor aún, una **Serie** de nuestro Dataframe, que ya contiene esos nombres.

```python
plt.pie(df['cantidad'],
       labels=df['etiquetas']);
```

Para que nuestro gráfico muestre el **porcentaje** representado por cada porción, podemos manipular el parámetro `autopct`:

```python
plt.pie(df['cantidad'],
       labels=df['etiquetas'],
       autopct='%.1f%%');
```

El parámetro `colors` sirve para establecer qué **colores** queremos asignarles a las porciones de nuestro pastel:

```python
plt.pie(df['cantidad'],
       labels=df['etiquetas'],
       autopct='%1.1f%%',
       colors=['#87CEEB', '#6495ED', '#000080', '#40E0D0']);
```

Con `shadow` podemos hacer que nuestro gráfico proyecte una **sombra** sutil:

```python
plt.pie(df['cantidad'],
       labels=df['etiquetas'],
       autopct='%1.1f%%',
       colors=['#87CEEB', '#6495ED', '#000080', '#40E0D0'],
       shadow=True);
```

Este es interesante. Si quieres que algunas porciones de la torta **se separen del resto**, puedes establecer las "distancias" que cada porción va a mantener con respecto al centro de la torta, manippulando el mpetodo `explode`:

```python
plt.pie(df['cantidad'],
       labels=df['etiquetas'],
       autopct='%1.1f%%',
       colors=['#87CEEB', '#6495ED', '#000080', '#40E0D0'],
       shadow=True,
       explode=(0, 0, 0, 0.5));
```

Y por último, me gustarpia que mi gráfico tenga un **título**. Esto no implica manipular elementos de `pie()`, sino agregar este elemento a través de un método de `plt` llamado `title()`. Esto significa que sin importar el tipo de gráfico que estemos creando, siempre podremos agregar títulos con el método `plt.title()`:

```python
plt.pie(df['cantidad'],
       labels=df['etiquetas'],
       autopct='%1.1f%%',
       colors=['#87CEEB', '#6495ED', '#000080', '#40E0D0'],
       shadow=True,
       explode=(0, 0, 0, 0.5))
plt.title("Mis frutas");
```
