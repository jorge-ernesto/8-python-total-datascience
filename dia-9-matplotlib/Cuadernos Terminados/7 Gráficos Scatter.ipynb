# Gráficos Scatter

La palabra *Scatter* significa algo así como *desparramar*. Entonces los **scatter plots** (o *gráficos scatter*), se llaman así porque lo que hacen es tomar todos los registros individuales que hay en nuestros datos y desparramarlos como puntos en una cuadrícula, como puedes ver en [este ejemplo](https://matplotlib.org/stable/plot_types/basic/scatter_plot.html#sphx-glr-plot-types-basic-scatter-plot-py) que nos muestra el sitio oficial de Matplotlib.

Vamos a crear nuestros propios gráficos de dos formas diferentes:
- con el método `plot()` (que ya hemos usado para crear gráficos de líneas)
- con el método `scatter()` (que es más dedicado)

Primero vamos a crear un conjunto de datos, y luego lo vamos a representar visualmente con Matplotlib.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
```

### `plot()`

Veamos cómo se ve esto con el método `plot()`.

```python
help(plt.plot)
```

```python
plt.plot(x, y);
```

```python
plt.plot(x, y, 'o');
```

```python
plt.plot(x, y, 'ro-');
```

```python
# np.random.seed(0)  # Establecer semilla de forma global en Numpy
aleatorio = np.random.RandomState(0)  # Establecer semilla de forma independiente en Numpy
marcadores = ['o', 'x', '+', 'v', 's', 'd']
```

```python
aleatorio.rand(5)
```

```python
aleatorio.rand(5)
```

```python
for marcador in marcadores:
    plt.plot(aleatorio.rand(5), aleatorio.rand(5), marcador)
```

```python
plt.plot(x, y, 'bo:');
```

```python
plt.plot(x, y, 'p-');
```

```python
plt.plot(x, y, 'p-', color='red');
```

```python
plt.plot(x, y, 'p', color='red', markersize=15);
```

```python
plt.plot(x, y, 'p-', color='red', markersize=15, linewidth=4);
```

```python
plt.plot(x, y, 'p-', color='red', markersize=15, linewidth=4, markerfacecolor='white');
```

### `scatter()`

Veamos cómo se ve esto con el método `scatter()`.

```python
help(plt.scatter)
```

```python
plt.scatter(x, y);
```

```python
# np.random.seed(0)  # Establecer semilla de forma global en Numpy
aleatorio = np.random.RandomState(0)  # Establecer semilla de forma independiente en Numpy
plt.scatter(aleatorio.randn(100),
           aleatorio.randn(100),
           c=aleatorio.randn(100),  # Establecer colores basado en 100 valores aleatorios
           s=aleatorio.randn(100) * 1000,  # Establecer tamaño basado en 100 valores aleatorios
           alpha=0.3,  # Transparencia
           cmap='viridis')  # Gama de colores: https://matplotlib.org/stable/users/explain/colors/colormaps.html#sequential
plt.colorbar();
```
